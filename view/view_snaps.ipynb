```python
import nyroglancer
import h5py
import os
import numpy as np


snap = '/groups/turaga/home/ecksteinn/circle_cvae/models/snapshots/run_0/snap_58000.h5'

f0 = h5py.File(snap, 'r')

x = f0['x']
y = f0['y']
y_out = f0['y_out']

voxel_size=(1,1,1)

viewer = nyroglancer.Viewer()
viewer.add(x, name='x', voxel_size=voxel_size)
viewer.add(np.array(y, dtype=np.float32), name='y', voxel_size=voxel_size)
viewer.add(np.array(y_out, dtype=np.float32), name='y_out', voxel_size=voxel_size)

viewer.set_hostname('localhost:8800')
viewer.show()
#f0.close()
```

```python
#Plot z if dim == 2
import matplotlib.pyplot as plt
import h5py
import numpy as np

snap = '/groups/turaga/home/ecksteinn/circle_cvae/models/snapshots/run_0/snap_44000.h5'
f0 = h5py.File(snap, 'r')

z = f0['z']
y = f0['y']
off = [np.array(np.where(np.array(y[i,:])>0.5)[0][0], 
                np.where(np.array(y[i,:])>0.5)[1][0]) -\
                np.array([28/2,28/2])for i in range(np.shape(y)[0])]

c = [np.linalg.norm(k) for k in off]

plt.scatter(z[:,0], z[:,1], c=c)
f0.close()
```
